## Standard optimization problem
Let us initially consider a standard optimization problem. So, suppose we want to find a vector $\mathbf{x} \in \mathbb{R}^n$ that minimizes a given functional $f(\mathbf{x})$. The optimization problem can be formulated as follows:

$$minimize \quad f(\mathbf{x})$$
subject to a constraint $\ldots$

$f(\mathbf{x})$ is referred with different names depending on the context. For example, in machine learning, it is called the loss function. In engineering, it is called the objective function or cost function. The constraint can be an equality constraint, an inequality constraint, or a combination of both. To exemplify, let us consider the following optimization problem:

$$\underset{\mathbf{x}}{min} \, 0.9 x_1^2 - 0.4 x_1 x_2 + 0.6 x_2^2 - 6.4 x_1 - 0.8 x_2$$
subject to $- 1 \leq x_1 \leq 2, 0 \leq x_2 \leq 3$.

The objective function is then $f(\mathbf{x}) = 0.9 x_1^2 - 0.4 x_1 x_2 + 0.6 x_2^2 - 6.4 x_1 - 0.8 x_2$. The constraint is the box constraint $- 1 \leq x_1 \leq 2, 0 \leq x_2 \leq 3$. The goal is to find the vector $\mathbf{x} = [x_1, x_2]$ that minimizes the objective function $f(\mathbf{x})$ while satisfying the constraint.

Great! Now, let us talk about vector space. Notice in the above optimization problem seeks to find a vector $\mathbf{x} \in \mathbb{R}^2$. What is $\mathbb{R}^2$? $\mathbb{R}^2$ is the set of all vectors with two components, i.e., $\mathbb{R}^2 = \{[x_1, x_2] \, | \, x_1, x_2 \in \mathbb{R}\}$. The set $\mathbb{R}^2$ is an example of a vector space.

Again, with a less informal definition, notice that two vectors $\mathbf{x}$ and $\mathbf{y}$ in $\mathbb{R}^2$ can be added to form a new vector $\mathbf{z} = \mathbf{x} + \mathbf{y}$. The vector $\mathbf{z}$ is also in $\mathbb{R}^2$. Also, consider a vector $y = \alpha \mathbf{x}$, where $\alpha$ is a scalar. The vector $y$ is also in $\mathbb{R}^2$. These two properties are essential to define a vector space. There are other properties as direct consequence of these two properties. If you want to know more about vector space, I recommend reading a book on linear algebra.
